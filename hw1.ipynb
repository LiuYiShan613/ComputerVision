```python
nr = 512
nc = 512
nchan = 1
```

讀取整張圖

```python
def raw_read(imgData):
    c = 0 #用 c來記錄 raw檔讀取到的 index位置
    img = np.zeros((nr, nc, nchan))
    for i in range(nr):
        for j in range(nc):
            for k in range(0, nchan):
                img[i, j, k] = imgData[c]
                c=c+1
    return img
```

讀取中間 10*10 pixel值

```python
def raw_read2(imgData):
    c = 0
    img = np.zeros((nr, nc, nchan))
    for i in range(nr):
        for j in range(nc):
            for k in range(0, nchan):
                img[i, j, k] = imgData[c]
                c=c+1
    #取中間 10 pixel
    cropimg = img[251:261,251:261]
    return cropimg
```

# Log Transformations 

bmp 圖檔

```python
#複製3個 bmp圖檔用來做 Log Transformations
imgData4_log = imgData4.copy()
imgData5_log = imgData5.copy()
imgData6_log = imgData6.copy()
```

```python
import math

def Log(im, nchan):
    
    height = 512
    width = 512
    #find maximum pixel in image
    r = np.max(im)

    for row in range(height):
        for col in range(width):
            for chan in range(nchan):
                im[row][col][chan] = (int)((255/ math.log(1+r)) * math.log(1 + im[row][col][chan]))
            
    return im
```

# Gamma Transformation 

```python
#複製3個 bmp圖檔用來做 Gamma Transformation
imgData4_gamma = imgData4.copy()
imgData5_gamma = imgData5.copy()
imgData6_gamma = imgData6.copy()
```

bmp圖檔

```python
def ga(im, gamma, nchan):
    height = 512
    width = 512
    for row in range(height):
        for col in range(width):
            for chan in range(nchan):
                im[row][col][chan] = (int)(255*(im[row][col][chan] / 255) ** gamma)
    return im
```

以下皆用gamma=3為例

# image negative

```python
# input image and the right image channel
def negative(im, nchan):
    height = 512
    width = 512
    for row in range(height):
        for col in range(width):
            for chan in range(nchan):
                im[row][col][chan] =  255 - im[row][col][chan]        
    return im
```

# Nearest Neighbor interpolation

以boat圖檔為例

```python
img = cv2.imread('boat.bmp')
```

```python
def nearest_neighbor_resize(img, new_h, new_w):
    # height and width of the input img
    h, w = img.shape[0], img.shape[1]
    # new image with rgb channel
    ret_img = np.zeros(shape=(new_h, new_w, 3), dtype='uint8')
    # scale factor
    s_h, s_c = (h * 1.0) / new_h, (w * 1.0) / new_w

    # insert pixel to the new img
    for i in range(new_h):
        for j in range(new_w):
            p_x = int(j * s_c)
            p_y = int(i * s_h)

            ret_img[i, j] = img[p_y, p_x]

    return ret_img
```

1. (512 X 512) -> (128 X 128)

```python
img_1 = nearest_neighbor_resize(img.copy(), 128, 128)
```

```python
plt.imshow(img_1)
```

2. (512 X 512) -> (32 X 32)

```python
img_2 = nearest_neighbor_resize(img.copy(), 32, 32)
```

```python
plt.imshow(img_2)
```

3. (32x32) -> (512x512)

```python
#用上面 32*32的img_2當輸入圖片
img_3 = nearest_neighbor_resize(img_2.copy(), 512, 512)
```

```python
plt.imshow(img_3)
```

4. (512x512) -> (1024x512)

# Bilinear interpolation

以baboon圖檔為例

```python
imgbi = cv2.imread('baboon.bmp')
```

```python
def GetBilinearPixel(imArr, posX, posY):
    out = []
    
    modXi = int(posX)
    modYi = int(posY)
    modXf = posX - modXi
    modYf = posY - modYi
    #正常情況取 modXi+1/modYi+1的值，除非超過圖片邊緣的值，則以圖片最邊緣的值為主
    modXiPlusOneLim = min(modXi+1, imArr.shape[1]-1)
    modYiPlusOneLim = min(modYi+1, imArr.shape[0]-1)
    
    # pixels in four corners
    for chan in range(imArr.shape[2]):
        bl = imArr[modXi, modYi, chan]
        br = imArr[modXi, modYiPlusOneLim, chan]
        tl = imArr[modXiPlusOneLim, modYi, chan]
        tr = imArr[modXiPlusOneLim, modYiPlusOneLim, chan]
        # calculate interpolation
        b = modXf * br + (1. - modXf) * bl
        t = modXf * tr + (1. - modXf) * tl
        pxf = modYf * t + (1. - modYf) * b
        out.append(int(pxf+0.5))
        
    return out
```

```python
#find position in original image
def bili(img, new_h, new_w):
    resizeImg = np.zeros(shape=(new_h, new_w, 3), dtype='uint8')
    rowScale = float(img.shape[0]) / float(resizeImg.shape[0])
    colScale = float(img.shape[1]) / float(resizeImg.shape[1])
    
    for r in range(new_h):
        for c in range(new_w):
            orir = r * rowScale
            oric = c * colScale
            resizeImg[r, c] = GetBilinearPixel(img, orir, oric)
            
    return resizeImg
```

1. (512 X 512) -> (128 X 128)

```python
nimgbi1 = bili(imgbi.copy(), 128, 128)
plt.imshow(nimgbi1)
```

2. (512 X 512) -> (32 X 32)

```python
nimgbi2 = bili(imgbi.copy(), 32, 32)
plt.imshow(nimgbi2)
```

3. (32x32) -> (512x512)

```python
#用上面 32*32的 nimgbi2當輸入圖片
nimgbi3 = bili(nimgbi2.copy(), 512, 512)
plt.imshow(nimgbi3)
```

4. (512x512) -> (1024x512)

```python
nimgbi4 = bili(imgbi.copy(), 1024, 512)
plt.imshow(nimgbi4)
```

```python
#用上面 128*128的 nimgbi1當輸入圖片
nimgbi5 = bili(nimgbi1.copy(), 256, 512)
plt.imshow(nimgbi5)
```
